```python
import pandas as pd
```

```python
cols_to_use = ['FID', 'IID', 'original_gldm_LargeDependenceHighGrayLevelEmphasis', 'original_shape_Sphericity']
```

```python
df = pd.read_csv('radiomics_spleen.csv')
df_new = df.drop_duplicates('ID').dropna()
```

```python
df_new.head()
```

```python
df_new['FID'] = df_new.ID
df_new['IID'] = df_new.ID
gwas_pheno = df_new[cols_to_use]
print(gwas_pheno.shape)
```

```python
gwas_pheno = gwas_pheno.rename(columns = {'original_gldm_LargeDependenceHighGrayLevelEmphasis': 'grayleveldependence', 'original_shape_Sphericity':'sphericity'})
gwas_pheno.head()
```

```python
gwas_pheno.to_csv('gwas_pheno_small_spleen.csv')
```
