```python
import pandas as pd
```

```python
cols_to_use = ['FID', 'IID']
spleen_prev_cols = ['original_glszm_GrayLevelNonUniformity', 'original_glcm_Id', 'original_shape_Sphericity']
spleen_inc_cols = ['original_firstorder_Kurtosis', 'original_gldm_LargeDependenceHighGrayLevelEmphasis', 
                  'original_glszm_GrayLevelNonUniformity', 'original_glszm_SmallAreaLowGrayLevelEmphasis',
                   'original_shape_Maximum2DDiameterSlice']
spleen_col_map = {'original_glszm_GrayLevelNonUniformity': 'graynonuniform',
                  'original_glcm_Id': 'grayinversediff',
                  'original_shape_Sphericity':'sphericity',
                  'original_firstorder_Kurtosis':'kurtosis',
                  'original_gldm_LargeDependenceHighGrayLevelEmphasis': 'graydependence', 
                  'original_glszm_SmallAreaLowGrayLevelEmphasis': 'grayemphasis',
                  'original_shape_Maximum2DDiameterSlice': 'max2Ddiam',
                  }
```

```python
df = pd.read_csv('radiomics_spleen.csv')
df_new = df.drop_duplicates('ID').dropna()
```

```python
df_new.head()
```

```python
df_new['FID'] = df_new.ID
df_new['IID'] = df_new.ID
gwas_pheno = df_new[cols_to_use+spleen_prev_cols+spleen_inc_cols]
gwas_pheno = gwas_pheno.loc[:,~gwas_pheno.columns.duplicated()]
print(gwas_pheno.shape)
```

```python
gwas_pheno = gwas_pheno.rename(columns = spleen_col_map)
gwas_pheno.head()
```

```python
gwas_pheno.to_csv('gwas_pheno_spleen.csv')
```
